```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_digits

#load dataset digits
digits = load_digits()

#calculate distribution of the sum of each digit
unique, counts = np.unique(digits.target, return_counts=True)

#plot figure
plt.figure(figsize=(8, 5))
sns.barplot(x=unique, y=counts, palette='coolwarm')

#add labels and titles
plt.xlabel("Digit (0-9)")
plt.ylabel("Jumlah Data")
plt.title("Data in the Digits Dataset")
#displays all label digits
plt.xticks(unique)
plt.show()


```

```python
#divide the dataset into training data and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
```

```python
#train a random forest model on the dataset
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
#testing the performance of the Random Forest model after making predictions on test data
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = rf.predict(X_test)

rf_accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:' ,rf_accuracy)
```

```python
#visualize the Confusion Matrix table
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
            xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
#displays the model evaluation in tabular form
print(classification_report(y_test,y_pred, target_names=[str(i) for i in digits.target_names]))
```

```python
#train a naive bayes model on the dataset
from sklearn.naive_bayes import GaussianNB
```

```python
#testing the performance naive bayes model after making predictions on test data
nb_model = GaussianNB()
nb_model.fit(X_train,y_train)

y_pred = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test,y_pred)

print('Accuracy : ', nb_accuracy)
```

```python
#visualize the Confusion Matrix table
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True, fmt='d',cmap='Blues', xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.show()
```

```python
#displays the model evaluation in tabular form
print(classification_report(y_test,y_pred, target_names=[str(i) for i in digits.target_names]))
```

```python
#train a SVM model on the dataset
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
#testing the performance of SVM after making predictions on test data
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

svm_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', svm_accuracy)
```

```python
#visualize the Confusion Matrix table
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
            xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
#displays the model evaluation in tabular form
print(classification_report(y_test,y_pred, target_names=[str(i) for i in digits.target_names]))
```

```python
#algorithm comparison
models = ['Random Forest','Naive Bayes','SVM']
accuracies = [rf_accuracy,nb_accuracy,svm_accuracy]


plt.figure(figsize=(8,4))
bars = plt.bar(models, accuracies, color='purple')

plt.title('accuracy data set of digits (classification).', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Akurasi', fontsize=12)
plt.ylim(0, 1)


for bar, accuracy in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width() / 1, bar.get_height() + 0.01,
             f'{accuracy:.2f}', ha='center', va='bottom', color='black', fontsize=10)

plt.show()

```
